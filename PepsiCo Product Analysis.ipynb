```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('PepsiCo.xlsx')
```

```python
df
```

```python
# Basic information about the dataset
print(df.info())
```

```python
# Summary statistics of numeric columns
print(df.describe())
```

```python
# Checking for missing values
print(df.isnull().sum())
```

```python
# Distribution of products in different categories
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df)
plt.xticks(rotation=60)
plt.title('Distribution of Products in Different Categories')
plt.show()
```

```python
# Product launch status represented as a pie chart
plt.figure(figsize=(6, 6))
df['Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Product Launch Status')
plt.show()
```

```python
# Number of products launched over the years using a count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Year Launched', data=df, palette='winter')
plt.xticks(rotation=90)
plt.title('Number of Products Launched Over the Years')
plt.show()
```

```python
# Ownership distribution using a count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Ownership', data=df, palette='bwr')
plt.xticks(rotation=45)
plt.title('Ownership Distribution')
plt.show()
```

```python
# Relationship between the year launched and the region served using a scatter plot
plt.figure(figsize=(20, 10))
sns.scatterplot(x='Year Launched', y='Region Served', data=df, hue='Category', size='Ownership', palette='bwr')
plt.title('Year Launched vs. Region Served')
plt.show()
```

```python
grouped_data = df.groupby(['Year Launched', 'Region Served'])['Product Name'].count().reset_index()

# Grouped bar chart to visualize the relationship between Year Launched and Region Served
plt.figure(figsize=(16, 6))
sns.barplot(x='Year Launched', y='Product Name', hue='Region Served', data=grouped_data, palette='winter')
plt.xlabel('Year Launched')
plt.ylabel('Number of Products')
plt.title('Number of Products Launched by Year and Region')
plt.legend(title='Region Served')
plt.xticks(rotation=45)
plt.show()
```

```python
# Stacked bar chart to show the distribution of the "Status" attribute for each "Category."
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', hue='Status', data=df, palette='gnuplot2')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Status Distribution by Category')
plt.legend(title='Status', loc='upper right')
plt.xticks(rotation=75)
plt.show()
```

```python
# box plot to visualize the distribution of product counts for each "Region Served."
plt.figure(figsize=(10, 6))
sns.countplot(x='Region Served', data=df, palette='PuBu')
plt.xlabel('Region Served')
plt.ylabel('Product Count')
plt.title('Distribution of Product Counts by Region Served')
plt.xticks(rotation=45)
plt.show()
```

------ Key Insights from the Project ------
1. The dataset contains products from various categories, with 'Beverages' having the highest number of products.
2. The majority of products in the dataset are in the 'Launched' status, with fewer in 'Discontinued' and 'In Development' statuses.
3. The number of product launches has fluctuated over the years, with certain years experiencing higher launch rates.
4. Different entities own products in the dataset, with varying numbers of products attributed to each owner.
5. The scatter plot highlights the relationship between year launched and region served, categorized by product category and ownership.
6. The grouped bar chart shows the number of products launched in each year for different regions, providing insights into regional trends.
7. The stacked bar chart compares the status distribution within each category, showing the proportion of launched, discontinued, and in-development products.
8. The box plot indicates the distribution of product counts across different regions served, identifying variations in product presence.
